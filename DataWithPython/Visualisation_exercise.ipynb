```python
# Import packaging
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Histogram 

```python
# Importing the tips dataframe
tips = sns.load_dataset("tips")
tips
```

```python
# Plotting histogram for tips data
sns.histplot(data=tips, x="total_bill", binwidth=5, bins=30, kde=True, color='red')

#Title
plt.title('Total Bill ', weight = 'bold', fontsize = 10)
```

```python
# Plotting histogram for tips data
sns.histplot(data=tips, x="day", hue="sex", multiple="dodge", shrink=.8, stat='percent')

#Title
plt.title('Percentage of male and female', weight = 'bold', fontsize = 10)
```

```python
sns.histplot(data=tips, x="total_bill", binwidth=5, bins=30, kde=True, hue='smoker', multiple="stack", color='rocket', stat='percent')


#Title
plt.title('Percentage of smokers and Non-smokers', weight = 'bold', fontsize = 10)
```

#### Scatterplot

```python
# Importing the iris dataframe
iris = sns.load_dataset("iris")
iris
```

```python
#Plotting the scatterplot
sns.lmplot(x="sepal_length", y="petal_length", data=iris, ci=None)

#Title
plt.title('Sepal_length vs Petal_length', weight = 'bold', fontsize = 10)
```

```python
#Plotting the scatterplot with line of best fit for each species
sns.lmplot(x="sepal_length", y="petal_length", data=iris, hue='species', ci=None)

#Giving title to the bar chart
plt.title('Sepal_length vs Petal_length', weight = 'bold', fontsize = 10)
```

#### Treemap 

treemap 'titanic' dataset based on the class column (use matplotlib and squarify)
pie chart titanic survivors data

```python
import squarify
```

```python
# Importing the iris dataframe
titanic = sns.load_dataset("titanic")
titanic
```

pie chart titanic survivors data

```python
#Defining ratio gap
explode = (0.05, 0.05, 0.05)

#Labelling pie chart
labels= ['Male', 'Female']

#Defining pie chart colors
colors = ['mediumturquoise', 'darkcyan']

titanic["survived"].value_counts()
titanic.survived.value_counts().plot(kind="pie", y='',autopct='%1.0f%%', colors=colors, shadow=True, labels=None)

#Adding legends
plt.legend(bbox_to_anchor=(1,1), labels=labels)

#Title for the piechart
plt.title('Percentage of survivors', fontsize=16, weight = 'bold')
```

```python
# Group the dataset by Pclass and Survived and then unstack them
group = titanic.groupby(['class', 'survived'])
class_survived = group.size().unstack()
 
# Heatmap - Color encoded 2D representation of data.
sns.heatmap(class_survived, annot = True, fmt ="d")
```

```python
# Group the dataset by Pclass and Survived and then unstack them
group = titanic.groupby(['pclass', 'survived'])
pclass_survived = group.size().unstack()
squarify.plot(pclass_survived, annot = True, fmt ="d")
```
