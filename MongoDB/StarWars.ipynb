```python
import pymongo
client = pymongo.MongoClient()
db = client['starwars']
print(db)
```

```python
# Grabbing all the droid species - using the filter bracket{}
characters = db.characters.find({})
character_name = db.characters.find({"species.name":"Droid"}, {"name":1, "species.name":1, "_id":0})
for person in character_name:   
    print(person)
```

```python
# Find the height of Darth Vader, only return results for the name and the height
characters = db.characters.find({"name": "Darth Vader"}, {"name": 1, "height": 1, "_id": 0})
for document in characters:
    print(document)
```

```python
# Find all characters with yellow eyes, only return results for the names of the characters
yellow = db.characters.find({"eye_color":"yellow"}, {"name":1, "_id":0})
for y in yellow:
    print(y)
```

```python
# Find male characters. Limit your results to only show the first 3.
male = db.characters.find({"gender":"male"}, {"name":1, "_id":0}).limit(3)
for m in male:
    print(m)
```

```python
# Find the names of all the humans whose homeworld is Alderaan.
alderan = db.characters.find({"homeworld.name": "Alderaan", "species.name": "Human"}, {"name":1, "_id":0})
for document in alderan:
    print(document)
```

```python
# # Find the names of all the humans whose homeworld is Alderaan.
Alderaan = db.characters.find({
    "$and":[
    {"species.name": "Human"},
    {"homeworld.name": "Alderaan"}]},
    {"name":1, "_id": 0})
for a in Alderaan:
    print(a)
```

Aggregation & Updating the Data

```python
all_characters = db.characters.find({}, {"height":1, "mass":1, "_id":0})
for character in all_characters:
    print(character)
```

Updating the data

```python
# Removing unknown masses 
db.characters.update_many({"height":"unknown"}, {"$unset":{"height":""}})

db.characters.update_many({"mass":"unknown"}, {"$unset":{"mass":""}})
all_characters = db.characters.find({}, {"height":1, "mass":1, "_id":0})

for character in all_characters:
    print(character)
```

```python
# set height field to int version of original  --> NOT s string
db.characters.update_many({}, [{"$set":{"height":{"$toInt": "$height"}}}])

all_characters = db.characters.find({}, {"height":1, "mass":1, "_id":0})
for character in all_characters:
    print(character)
```

```python
## Sort out the comma problem
db.characters.update_one({"mass":"1,358"}, {"$set": {"mass":1358}})
all_characters = db.characters.find({}, {"height":1, "mass":1, "_id":0})
for character in all_characters:
    print(character)
```

```python
# Converting all the masses to double
db.characters.update_many({},[{"$set":{"mass":{"$toDouble": "$mass"}}}])

all_characters = db.characters.find({}, {"height":1, "mass":1, "_id":0})
for character in all_characters:
    print(character)
```

```python
# Average height of femala characters 
avg_female_height = db.characters.aggregate([{"$match":{"gender":"female"}}, {"$group":{"_id":"$gender", "avg_height":{"$avg":"$height"}}}])
for character in avg_female_height:
    print(character)
```
