```python
# Importing packages
import boto3
from pprint import pprint as pp
import json # importing json
from io import StringIO
import pandas as pd
```

```python
# Creating client object
s3_client = boto3.client('s3')
# Identifying bucket name
bucket_name = 'data-eng-resources'
```

3 different fish market files
- key1: 'python/fish-market-mon.csv'
- key2: 'python/fish-market-tues.csv'
- key3: 'python/fish-market.csv'

1. Read the fish market CSV data from all fish files


```python
# Reading the fish-market-mon.csv file
key = 'python/fish-market-mon.csv'
csv_obj = s3_client.get_object(Bucket=bucket_name, Key=key)
strbody  = csv_obj['Body']
market_mon_body = strbody.read().decode('utf-8') # gets the body and removes the binary string
```

```python
pp(market_mon_body)
```

```python
market_mon = pd.read_csv(StringIO(market_mon_body))
market_mon
```

```python
# Reading the fish-market-tues.csv file
key = 'python/fish-market-tues.csv'
csv_obj = s3_client.get_object(Bucket=bucket_name, Key=key)
strbody  = csv_obj['Body']
market_tue_body = strbody.read().decode('utf-8') # gets the body and removes the binary string
```

```python
market_tues= pd.read_csv(StringIO(market_tue_body))
market_tues
```

```python
# Reading the fish-market.csv file
key = 'python/fish-market.csv'
csv_obj = s3_client.get_object(Bucket=bucket_name, Key=key)
strbody  = csv_obj['Body']
fish_market_body = strbody.read().decode('utf-8') # gets the body and removes the binary string
```

```python
fish_market= pd.read_csv(StringIO(fish_market_body))
fish_market
```

Merging the 3 dataframes

```python
dataframes = [market_mon,market_tues, fish_market]
merging_dataframes = pd.concat(dataframes)
merging_dataframes
```

2. Transformation: data averaged by fish species

```python
# Average of all species 
transformed_data = merging_dataframes.groupby("Species").mean("Species")
transformed_data
```

```python
# Converting the transformed data into a csv file
transformed_data.to_csv('Transformed_Data.csv', index = True)
```
