Cleaning the dataframe

```python
# check for any duplicate 
json_df = json_df.dropDuplicates()
json_df.count()

''' We see that there are no duplicates within the dataframe'''
```

```python
# Identifying the null values with the dataframe 
json_df.select([count(when(col(i).isNull(),i)).alias(i) for i in json_df.columns]).display()
```

```python
# Calculating the % of null values in each column to determine which columns should be dropped 
for i in json_df.columns:
    no_null = json_df.filter(col(i).isNull()).count()
    total = json_df.count()
    percent_null = no_null / total * 100
    print(f"{i}: {percent_null:.2f}%")
```

```python
''' 
Looking at the % of null values in the dataframe, we can determine which columns contains the most null  values. 
Overall we can see that dvdCountry has the highest % of null values
There are allot of unrelated records in the dataframe, those should be taken into consideration
'''
# From the dataframe, we dropped _links, externals,image,summary,url,weight, ended and updated due to them being unrelated records to the question being asked
# The columns dvdCountry,externals,officialSite and webChannel were dropped as they contained high % of null values

json_df = json_df.drop("_links","dvdCountry","externals","image","officialSite","summary","url","webChannel","weight","updated","ended")

# Displaying the dataframe after dropping the columns
json_df.display()
```
